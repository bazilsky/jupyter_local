```python
import os 
import random 
import numpy as np 
import pandas as pd
import tensorflow as tf 
from pprint import pprint
from pylab import plt,mpl
plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'
pd.set_option('precision',4)
np.set_printoptions(suppress = True, precision = 4)
os.environ['PYTHONHASHSEED'] = '0'

```

```python
def set_seeds(seed = 100):
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)

set_seeds()
```

```python
def transform(x):
    y = 0.05 * x ** 2+ 0.2 * x + np.sin(x) + 5 
    y += np.random.standard_normal(len(x)) * 0.2
    return y

x = np.linspace(-2*np.pi, 2 * np.pi, 500)
a = transform(x)

plt.figure(figsize = (10,6))
plt.plot(x,a)
    
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator

```

```python
a = a.reshape((len(a),-1))
a[:5]
```

```python
from keras.models import Sequential
from keras.layers import SimpleRNN, LSTM, Dense
```

```python
lags = 5

g = TimeseriesGenerator(a,a,length = lags, batch_size = 5)


model = Sequential()
model.add(SimpleRNN(500, activation='relu',
                    input_shape=(lags, 1)))
model.add(Dense(1, activation='linear'))
model.compile(optimizer='rmsprop', loss='mse',
                       metrics=['mae'])

#model = Sequential()
#model.add(SimpleRNN(100,activation = 'relu', input_shape = (lags,1)))
#model.add(Dense(1,activation = 'linear'))
#model.compile(optimizer = 'adagrad', loss = 'mse', metrics = ['mae'])
```

```python
model.summary()
```

```python
%%time
model.fit_generator(g, epochs=500, steps_per_epoch=10,
                             verbose=False)
```

```python
# the above model is trained between -2pi and 2pi
# now we try to use the model to predict out of sample 
#for values between -6pi and 6pi 
```

```python
x = np.linspace(-6*np.pi, 6* np.pi, 1000)
d = transform(x)
```

```python
g_ = TimeseriesGenerator(d,d,length = lags, batch_size = len(d))
f = list(g_)[0][0].reshape((len(d) - lags), lags, 1)
y = model.predict(f,verbose = False)

```

```python
plt.figure(figsize = (12,10))
plt.plot(x[lags:], d[lags:], label = 'data',alpha = 0.75)
plt.plot(x[lags:],y,'r.', label = 'pred', ms = 3)
plt.axvline(-2*np.pi, c = 'g', ls = '--')
plt.axvline(2*np.pi, c = 'g', ls = '--')
plt.text(-15,22,'out-of-sample',fontsize = 18)
plt.text(-2,22,'in-sample', fontsize = 18)
plt.text(10,22,'out-of-sample',fontsize = 18)
plt.legend(fontsize = 20)
```
