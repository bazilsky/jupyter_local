```python
import numpy as np 
import pandas as pd
from pylab import plt, mpl # pylab rather than matplotlib 
plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 300 
mpl.rcParams['font.family'] = 'serif' 
np.set_printoptions(precision=4, suppress=True)
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
x, y = make_blobs(n_samples=100, centers=4,random_state=500, cluster_std=1.25)
x1, y1 = make_blobs(n_samples=100, centers=4,random_state=500, cluster_std=1.25)

diff = y-y1;  # random state variable is a seed to the pseudo random generator, will generate the same set if that is fixed, therefore diff = 0

```

```python
model = KMeans(n_clusters = 4,random_state = 0);
model.fit(x);


y_ = model.predict(x) # clsutering algorithm prediction 
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x[:,0],x[:,1],c = y_,cmap = 'viridis')
plt.title('Kmeans clustering algo test')
```
