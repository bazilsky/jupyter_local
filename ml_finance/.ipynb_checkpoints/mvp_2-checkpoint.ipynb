```python
import numpy as np 
import pandas as pd 
from pylab import plt, mpl 
plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 500
mpl.rcParams['font.family'] = 'serif'
#np.set_printoptions(precision=5, suppress=True,formatter={'float': lambda x: f'{x:6.3f}'})
```

```python
url = 'http://hilpisch.com/aiif_eikon_eod_data.csv'
raw = pd.read_csv(url,index_col = 0, parse_dates = True).dropna()
```

```python
raw.info()
```

```python
symbols = ['AAPL.O','MSFT.O','INTC.O','AMZN.O','GLD']
#symbols = ['AAPL.O','MSFT.O','INTC.O', 'AMZN.O', 'GLD']
rets = np.log(raw[symbols] / raw[symbols].shift(1)).dropna() # ths shift shifts the array by one row
```

```python
# plot of the normalised financial time series
(raw[symbols]/raw[symbols].iloc[0]).plot(figsize=(12,8))
```

```python
raw[symbols].shift(1).dropna()
```

```python
weights = len(rets.columns) * [1/len(rets.columns)]
np.dot(rets.mean(),weights)*252
```

```python
# portfolio return function 
def port_return(rets,weights):
    a  = np.dot(rets.mean(),weights)*252
    return a

port_return(rets,weights)
```

```python
def port_volatility(rets,weights):
    return np.dot(weights, np.dot(rets.cov() * 252 , weights)) ** 0.5

port_volatility(rets, weights)
```

```python
def port_sharpe(rets,weights):
    return port_return(rets, weights) / port_volatility(rets, weights)

port_sharpe(rets,weights)
```

```python
# randomize the portfolio weights like in a monte carlo simulation
w = np.random.random((1000,len(symbols)))
print(np.shape(w.sum(axis = 1)))
w = (w.T/ w.sum(axis = 1)).T # normalising the weights with the sum and the transpose is so that matriz division can be done
print(np.shape(w.T))
```

```python
pvr = [(port_volatility(rets[symbols],weights),port_return(rets[symbols],weights)) for weights in w]
pvr = np.array(pvr)
```

```python
psr = pvr[:,1]/pvr[:,0] # sharpe ratio calculation
```

```python
plt.figure(figsize=(12, 8))
fig = plt.scatter(pvr[:, 0], pvr[:, 1],c=psr, cmap='coolwarm')
cb = plt.colorbar(fig)
cb.set_label('Sharpe ratio')
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.title('|'.join(symbols));
```
