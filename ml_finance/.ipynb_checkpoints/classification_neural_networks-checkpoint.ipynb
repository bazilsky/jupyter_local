```python
# using neural networks to tackle classification problems

import numpy as np 
import pandas as pd
from pylab import plt, mpl
import tensorflow
from sklearn.neural_network import MLPRegressor
from keras.layers import Dense
from keras.models import Sequential
```

```python
f = 5 
n = 10 
np.random.seed(100)
```

```python
x = np.random.randint(0,2,(n,f)) # get random array and random.seed helps fix the randomness
y = np.random.randint(0,2,n)
```

```python
model = Sequential()
model.add(Dense(256,activation = 'relu',input_dim = f)) # input_dim refers to the 5 features that are the input
model.add(Dense(1,activation = 'linear'))
model.compile(loss = 'binary_crossentropy',optimizer = 'rmsprop',metrics = ['acc']) # what are metrics?
model.fit(x,y,epochs = 50, verbose = False)
```

```python
from keras.layers import Dense
from keras.models import Sequential
model = Sequential()
model.add(Dense(256,activation = 'relu',input_dim = 1)) # adding a rectilinear unit for activation 
model.add(Dense(1,activation = 'linear'))
model.compile(loss = 'mse',optimizer = 'rmsprop')
plt.figure(figsize = (10,8))
plt.plot(x,y,'ro',label = 'sample data')
for _ in range(1,6):
    model.fit(x,y,epochs = 100,verbose = False) # every loop the model is trained with 100 epochs 
    y_ = model.predict(x)
    mse = ((y-y_.flatten())**2).mean()  # flatten function used to make 2D array into 1D array by flattening
    print(f'round = {_} || mse = {mse:.3f}')
    plt.plot(x,y_,'--',label = f'round = {_}')
plt.legend()
```

```python
a = np.random.seed(0) # this statement makes a random number predictable
x = np.linspace(-1,1)
y = np.random.random(len(x))*2-1;
```

```python
plt.figure(figsize = (10,8))
plt.plot(x,y,'ro',label = 'sample data')
model = Sequential()
model.add(Dense(256,activation = 'relu',input_dim = 1))
for _ in range(3):
    model.add(Dense(256,activation = 'relu')) # keep adding a new activation layer for loop run
model.add(Dense(1,activation = 'linear'))
model.compile(loss = 'mse',optimizer = 'rmsprop')

for _ in range(1,8):
    model.fit(x,y,epochs = 500,verbose = False)
    y_ = model.predict(x)
    plt.plot(x,y_,'--',label = f'round = {_}')
    
plt.legend()

    

```
